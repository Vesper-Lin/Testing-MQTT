```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("stats_final.csv")
```

```python
fig, ax1 = plt.subplots(figsize=(12, 5))

sns.barplot(x='delay', y='avg_rate', hue='qos', data=df, ax=ax1, palette = 'rainbow')
sns.despine(fig)
```

```python
fig, ax2 = plt.subplots(figsize=(12, 5))

sns.barplot(x='delay', y='loss_rate', hue='qos', data=df, ax=ax2, palette = 'rainbow')
sns.despine(fig)
```

```python
fig, ax3 = plt.subplots(figsize=(12, 5))

sns.barplot(x='delay', y='out_of_order_rate', hue='qos', data=df, ax=ax3, palette = 'rainbow')
sns.despine(fig)
```

```python
fig, ax4 = plt.subplots(figsize=(12, 5))

sns.barplot(x='delay', y='mean_inter_message_gap', hue='qos', data=df, ax=ax4, palette = 'rainbow')
sns.despine(fig)
```
